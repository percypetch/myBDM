```python
#############################################################################
#### This function is used for randomly modify your data DON'T TOUCH IT #####
#############################################################################
import random
def augment(df, id):
    rng = np.random.RandomState(id)
    colidx = list(range(0,11))
    rowidx = list(range(0,1309))
    colc = rng.choice(colidx, 2)
    for col in colc:
        rowc = rng.choice(rowidx,20)
        for row in rowc:
            if (rng.rand() > 0.5):
                df.iloc[row, col] = None
    df_new = df.rename(columns={'Survived': 'Label'})
    return df_new
```

### 2.1 Please prepare this dataset for create machine learning model https://raw.githubusercontent.com/plenoi/BDM/master/titanic.txt Please Change 4305033 to <b>your student-id</b> (5 Points)

```python
df = pd.read_csv('https://raw.githubusercontent.com/plenoi/BDM/master/titanic.txt', sep='\t', header = 0)
df = augment(df, 622110137) # Please Change 4305033 to your student id
df = df.set_index('Name')
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df_clean_column = df
df_clean_column = df_clean_column.drop(['Cabin','Ticket Number','Life Boat'],1)
df_clean_column.isnull().sum(axis=0)
```

```python
df_clean_column['Port of Embarkation'].unique()
```

```python
df_clean_column.shape
```

```python
df_clean = df_clean_column.dropna(0,'any')
df_clean = df_clean.dropna(1,'any')
df_clean
```

```python
df_clean.isnull().sum()
```

```python
df_clean.dtypes
```

```python
p_class, levels = pd.factorize(df_clean['Passenger Class'])
df_clean['Passenger Class'] = p_class
df_clean['Sex'] = pd.factorize(df_clean['Sex'])[0]
df_clean['Label'] = pd.factorize(df_clean['Label'])[0]
df_clean['Port of Embarkation'] = pd.factorize(df_clean['Port of Embarkation'])[0]
```

```python
df_clean
```

```python
df_clean['Label'].unique()
```

```python
y = df_clean['Label'].values
X = df_clean.drop(['Label'],axis = 1).values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=622110137)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
scaler.fit(X_train)
X_train_norm = scaler.transform(X_train)
X_test_norm = scaler.transform(X_test)
```

```python
X_train_norm
```

### 2.2 Use prepared dataset from 2.1 to create machine learning model, then evaluate model performance using Train/Test Accuracy and Confusion Matrix. Note: please compare performance of 3 or more models. (5 Points)

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

params = {'C': [4, 8, 16]}
clf = GridSearchCV(LogisticRegression(random_state=0, solver='liblinear'),params, cv=10)
clf.fit(X_train_norm, y_train)
print("Best params : " + str(clf.best_params_))
print("10CV accuracy : "+str(clf.best_score_*100))
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
y_predict = clf.predict(X_test_norm)
print("Test accuracy : "+str(sum(y_test == y_predict)/len(y_test)*100))
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
target_names = ['negative', 'positive']
C = confusion_matrix(y_test,y_predict) 
C = C / C.astype(np.float).sum(axis=1)*100
sns.heatmap(C, annot=True, fmt="f",cmap="GnBu",xticklabels=target_names, yticklabels=target_names)
plt.ylabel("True Label")
plt.xlabel("Predicted Label")
plt.show()
```

# Support Vector Machines

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
params = {
    'C' : [1,2,4,8,16,32], # High C = Overfitting
    'gamma' : [0.03125, 0.0625, 0.125, 0.25, 0.5, 1, 2, 4, 8, 16, 32] # High gamma = Overfitting
}
clf = GridSearchCV(SVC(),params, cv=10)
clf.fit(X_train_norm, y_train)
print("Best params : " + str(clf.best_params_))
print("10CV accuracy : "+str(clf.best_score_*100))
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
y_predict = clf.predict(X_test_norm)
target_names = ['negative', 'positive']
print("Test accuracy : "+str(sum(y_test == y_predict)/len(y_test)*100))
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
target_names = ['negative', 'positive']
C = confusion_matrix(y_test,y_predict) 
C = C / C.astype(np.float).sum(axis=1)*100
sns.heatmap(C, annot=True, fmt="f",cmap="GnBu",xticklabels=target_names, yticklabels=target_names)
plt.ylabel("True Label")
plt.xlabel("Predicted Label")
plt.show()
```
